```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.load("Data/X.npy")
Y = np.load("Data/Y.npy")
```

```python
print("X Shape\n", X.shape)
print("Y Shape\n", Y.shape)
print("X Data\n", X[:5, :])
print("Y Data\n", Y[:5])
```

```python
q = np.array([10, 3])
plt.scatter(q[0], q[1])
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.show()
```

```python
def dist(a1, a2):
    return np.sum((a1-a2)**2)**.5
```

```python
def knn(X, Y, q, k=5):
    m = X.shape[0]
    l = []
    
    for i in range(m):
        d = dist(q, X[i])
        l.append((d, Y[i]))
    
    l.sort()
    l = np.array(l[:k])
    l = l[:, 1]
    uniq, freq = np.unique(l, return_counts = True)
    p = np.argmax(freq)
    pred = uniq[p]
    return int(pred)
    
print(knn(X, Y, q))
```
